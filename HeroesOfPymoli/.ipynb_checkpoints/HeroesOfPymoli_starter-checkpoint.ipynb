```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
game_data = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(game_data)
purchase_data
```

## Player Count

* Display the total number of players


```python
player_count = purchase_data["SN"].value_counts()
len(player_count)
pretty_frame = pd.DataFrame({"Total Players":[len(player_count)]})
pretty_frame
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_items = purchase_data["Item Name"].unique()
number_of_unique_items = len(unique_items)


ave_price = purchase_data ["Price"].mean()


num_of_purchases = purchase_data["Purchase ID"].count()

total_revenue = purchase_data["Price"].sum()

data_frame = pd.DataFrame({"Number of Unique Items": [number_of_unique_items],
                        "Average Price": [ave_price], "Number of Purchases" : [num_of_purchases], 
                             "Total Revenue":  [total_revenue]                       
                          })
data_frame

data_frame.style.format({"Average Price": "${:,.2f}","Total Revenue": "${:,.2f}"})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python


#cleaning the data of duplicates
clean_data = purchase_data.drop_duplicates(['SN'], keep = 'first')
total_clean_number = len(clean_data)


#Count of male players 
male_purchasers =clean_data.loc[purchase_data["Gender"] == "Male", :]
male_purchasers_count = len(male_purchasers)


#Count of female players 
female_purchasers =clean_data.loc[purchase_data["Gender"] == "Female", :]
female_purchasers_count = len(female_purchasers)

#Count of Other / Non-Disclosed 
other_purchasers =clean_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]
other_purchasers_count = len(other_purchasers)

#Calculating percents
male_percent = (male_purchasers_count/total_clean_number)*100
female_percent = (female_purchasers_count/total_clean_number)*100
other_percent = (other_purchasers_count/total_clean_number)*100


#Creating output frame from data above
output_frame = {
    "Total Count": [male_purchasers_count, female_purchasers_count, other_purchasers_count],
    "Percentage of Players": [male_percent,female_percent,other_percent]
}

#Outputting frame
data_frame_2 = pd.DataFrame(output_frame)
data_frame_2.index = ['Male','Female','Other / Non-Disclosed']
data_frame_2

#styling frame
data_frame_2.style.format({"Percentage of Players": "{:,.2f}%"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#obtain purchase count for genders
#Total number of purchases done above = "num_of_purchases = purchase_data["Purchase ID"].count()""
male_purchasers_whole_count =purchase_data.loc[purchase_data["Gender"] == "Male", :]
purchase_count_male = len(male_purchasers_whole_count)

female_purchasers_whole_count =purchase_data.loc[purchase_data["Gender"] == "Female", :]
purchase_count_female = len(female_purchasers_whole_count)

other_purchasers_whole_count =purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed", :]
purchase_count_other = len(other_purchasers_whole_count)


#avg. purchase price
# ave_price = purchase_data ["Price"].mean()
male_ave_price = male_purchasers_whole_count ['Price'].mean()
female_ave_price = female_purchasers_whole_count ['Price'].mean()
other_ave_price = other_purchasers_whole_count ['Price'].mean()

#Total Purchase Value
male_total_price = male_purchasers_whole_count ['Price'].sum()
female_total_price = female_purchasers_whole_count ['Price'].sum()
other_total_price = other_purchasers_whole_count ['Price'].sum()

#avg. purchase total per person
male_ave_purchase = male_total_price/male_purchasers_count
female_ave_purchase = female_total_price/female_purchasers_count
other_ave_purchase = other_total_price/other_purchasers_count

#Creating output frame from data above
output_3_frame = {
    "Purchase Count": [purchase_count_male, purchase_count_female, purchase_count_other],
    "Average Purchase Price": [male_ave_price, female_ave_price, other_ave_price],
    "Total Purchase Value": [male_total_price,female_total_price,other_total_price],
    "Avg Total Purchase per Person": [male_ave_purchase,female_ave_purchase,other_ave_purchase]
}

#Outputting frame
data_frame_3 = pd.DataFrame(output_3_frame)
data_frame_3.index = ['Male','Female','Other / Non-Disclosed']
data_frame_3

 #styling frame
data_frame3.style.format({"Average Purchase Price": "${:,.2f}","Total Purchase Value": "${:,.2f}","Avg Total Purchase per Person": "${:,.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Categorize the existing players using the age bins.

bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 99]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

# clean_data has duplicates names removed form the list
pd.cut(clean_data["Age"], bins, labels=group_labels)
# adding age group to the clearn data frame 
clean_data["Age Group"] = pd.cut(clean_data["Age"], bins, labels=group_labels)

# creating total age count for groups 
less_then_10 =clean_data.loc[clean_data["Age Group"] == "<10", :]
less_then_10_count = len(less_then_10)

ten_14 = clean_data.loc[clean_data["Age Group"] == "10-14", :]
ten_14_count = len(ten_14)

fourteen_19 = clean_data.loc[clean_data["Age Group"] == "15-19", :]
fourteen_19_count = len(fourteen_19)

twenty_24 = clean_data.loc[clean_data["Age Group"] == "20-24", :]
twenty_24_count = len(twenty_24)

twentyfive_29 = clean_data.loc[clean_data["Age Group"] == "25-29", :]
twentyfive_29_count = len(twentyfive_29)

thirty_34 = clean_data.loc[clean_data["Age Group"] == "30-34", :]
thirty_34_count = len(thirty_34)

thirtyfive_39 = clean_data.loc[clean_data["Age Group"] == "35-39", :]
thirtyfive_39_count = len(thirtyfive_39)

more_then_40 = clean_data.loc[clean_data["Age Group"] == "40+", :]
more_then_40_count = len(more_then_40)

# creating percentage numbers for ages
percent_less_then_10 = (less_then_10_count/total_clean_number)*100
percent_ten_14 =(ten_14_count/total_clean_number)*100
percent_fourteen_19 =(fourteen_19_count/total_clean_number)*100
percent_twenty_24 =(twenty_24_count/total_clean_number)*100
percent_twentyfive_29 =(twentyfive_29_count/total_clean_number)*100
percent_thirty_34 =(thirty_34_count/total_clean_number)*100
percent_thirtyfive_39 =(thirtyfive_39_count/total_clean_number)*100
percent_more_then_40 =(more_then_40_count/total_clean_number)*100


#Creating output frame from data above
output_4_frame = {
    "Total Count": [less_then_10_count, ten_14_count, fourteen_19_count,twenty_24_count,
                    twentyfive_29_count,thirty_34_count,thirtyfive_39_count, more_then_40_count],
    "Percentage of Players": [percent_less_then_10, percent_ten_14, percent_fourteen_19, percent_twenty_24,
                              percent_twentyfive_29, percent_thirty_34, percent_thirtyfive_39,percent_more_then_40]
}

#Outputting frame
data_frame_4 = pd.DataFrame(output_4_frame)
data_frame_4.index = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
data_frame_4

 #styling frame
data_frame_4.style.format({"Percentage of Players": "{:,.2f}%"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns



* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Retrieve the Item ID, Item Name, and Item Price columns
most_popular_data = purchase_data[['Item ID','Item Name', 'Price']]


# Group by Item ID and Item Name. 
mp_group_data = most_popular_data.groupby(['Item ID','Item Name'])

#Perform calculations to obtain purchase count, average item price, and total purchase value
mp_purchase_count = mp_group_data['Price'].count()
mp_purchase_sum = mp_group_data['Price'].sum()
mp_purchase_price = mp_purchase_sum/mp_purchase_count


#Create a summary data frame to hold the results
#Outputting frame
data_frame_7 = pd.DataFrame({'Purchase Count':mp_purchase_count,
                             'Item Price':mp_purchase_price,
                             'Total Purchase Value':mp_purchase_sum})
                             
                                                    
#Sort the purchase count column in descending order                  
mp_formatting = data_frame_7.sort_values(['Purchase Count'], ascending=False).head()

#styling frame
mp_formatting.style.format({'Item Price': '${:,.2f}','Total Purchase Value':'${:,.2f}'})
```
